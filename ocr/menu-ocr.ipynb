Attempt to convert to B&W. This is usually good for ocr, we found that this made things worse on very small files

```python
from PIL import Image

column = Image.open('./yelp-photos-sample/22pCo74Z7ec-rECnVNO7DA.jpg')
gray = column.convert('L')
blackwhite = gray.point(lambda x: 0 if x < 200 else 255, '1')
blackwhite.save("./yelp-photos-sample/22pCo74Z7ec-rECnVNO7DA_bw.jpg")
```

Attempt to process the files and extract (better results not using B&W for this corpus)

```python
from tesserocr import PyTessBaseAPI

import sys

try:
    with PyTessBaseAPI() as api:
        api.SetImageFile('./yelp-photos-sample/22pCo74Z7ec-rECnVNO7DA.jpg')
        print(api.GetUTF8Text())
except:
    print(sys.exc_info()[0])
finally:
    print('')
```

The next step demonstrates batch processing

```python
from tesserocr import PyTessBaseAPI

import sys
import glob

to_process = glob.glob('./yelp-photos-sample/*')

print(len(to_process))

for image in to_process:
    try:
        with PyTessBaseAPI() as api:
            print(image)
            api.SetImageFile(image)
            print(api.GetUTF8Text())
    except:
        print(sys.exc_info()[0])
    finally:
        print('')
```
